```python
import pandas as pd
```

```python
# we convert date object datatype to datetime datatype
df=pd.read_csv("https://raw.githubusercontent.com/SwatiR20/Python_Amazon_Sales_Analysis/main/Amazon%20Sale%20Report.csv",parse_dates=["Date"])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# here we drop 2 column have whole column null value
df.drop(columns=["New","PendingS",],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# here we drop 2 column have whole column same value as india for ship country and for postal code column is not relevant for me
df.drop(columns=["ship-country","ship-postal-code"],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["fulfilled-by"],axis=1,inplace=True)
```

```python
df.head(2)
```

```python
df["B2B"].unique()
```

```python
# Drop rows where either "ship-city" or "ship-state" columns have null values
df = df.dropna(subset=["ship-city", "ship-state"])
```

```python
# drop rows where amount is null
df=df.dropna(subset=["Amount"],axis=0)
```

```python
df.isnull().sum()
```

```python
df["currency"].unique()
```

```python
df.drop(columns=["currency"],axis=1,inplace=True)
```

```python
# Group by "ship-city" and calculate the mean of the "Amount" column
mean_amount_by_city = df.groupby("ship-city")["Amount"].mean()
mean_amount_by_city
```

```python
import matplotlib.pyplot as plt

quantity_by_category = df.groupby("Category")["Qty"].sum().sort_values(ascending=False)
plt.figure(figsize=(5, 4))
x=quantity_by_category.plot(kind="bar")
x.bar_label(x.containers[0])
plt.show()
# here we can see that tshirt has been sales more in amazon and lowest sales is watch
```

```python
import matplotlib.pyplot as plt

quantity_by_category = df.groupby("Status")["Order ID"].count().sort_values(ascending=True)
plt.figure(figsize=(5, 4))
x=quantity_by_category.plot(kind="barh")
x.bar_label(x.containers[0])
plt.show()
# here we can see that product is shipped from amazon and product damaged is less = 1 means product is reach to buyer succesflully
```

```python
import matplotlib.pyplot as plt

quantity_by_category = df.groupby("Courier Status")["Order ID"].count().sort_values(ascending=True)
plt.figure(figsize=(5, 4))
x=quantity_by_category.plot(kind="barh")
x.bar_label(x.containers[0])
plt.show()
#here product is shipped to buyer is more than cancelling the product which is good sign for delivery and ontime serivces
```

```python
import matplotlib.pyplot as plt

quantity_by_category = df.groupby("ship-city")["Order ID"].count().sort_values(ascending=True).nlargest(3)
plt.figure(figsize=(5, 4))
x=quantity_by_category.plot(kind="barh")
x.bar_label(x.containers[0])
plt.show()
# here we see top 5 city order is generated from banglore city and less in chennai city
```

```python
import matplotlib.pyplot as plt

quantity_by_category = df.groupby("ship-state")["Order ID"].count().sort_values(ascending=True).nlargest(3)
plt.figure(figsize=(5, 4))
x=quantity_by_category.plot(kind="barh")
x.bar_label(x.containers[0])
plt.show()
# here we see top 3 state order is generated from maharashtra
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing order data

# Group by "ship-state" and count orders
quantity_by_category = df.groupby("Category")["Order ID"].count().sort_values(ascending=True).nlargest(3)

# Group by both "ship-state" and "size", and count orders for each size within each state
size_distribution = df.groupby(["Category", "Size"])["Order ID"].count().reset_index()

# Pivot the data to have "ship-state" as index, "size" as columns, and counts as values
size_pivot = size_distribution.pivot(index="Category", columns="Size", values="Order ID")

# Plotting
plt.figure(figsize=(10, 6))

# Plot the bar chart for quantity by category
ax = quantity_by_category.plot(kind="barh", color="lightblue", alpha=0.6, label="Total Orders")

# Plot the stacked bar chart for size distribution within each category
size_pivot.plot(kind="barh", stacked=True, ax=ax)

# Adjust labels and legend
plt.xlabel("Number of Orders")
plt.ylabel("Ship State")
plt.title("Orders Count and Size Distribution by Ship State")
plt.legend(title="Size")

plt.show()
```

```python
#lets crate heatmap now so to get idea where size of category matter in amount or not, does amount  increase or decrease with respect to category by size?
import seaborn as sns

grouped_data = df.groupby(['Category','Size'])['Amount'].mean().reset_index()
pivot_table = grouped_data.pivot(index='Category', columns='Size', values='Amount')
print(pivot_table)
```

```python
# here we create heatmap, blank value tell there is no data available for that category and  size
# light colour means have less relation ship between category and size with respect to price
# dark colour means have high relation ship between category and size with respect to price
# but from this data we see price does not increase with respect to increase in size
# heatmap we plot is not proper we just create to see relationship, always proper heatmap is seen when blank data is not present
plt.figure(figsize=(12,20))
sns.heatmap(pivot_table, cmap="YlGnBu",cbar_kws={'label': 'Mean Price'},annot=True,fmt=".2f")
```
